```python
import cv2
import pytesseract
import logging
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
basewidth = 1000
img = Image.open("image5.png")
img = img.convert('L')
wpercent = (basewidth / float(img.size[0]))
hsize = int((float(img.size[1]) * float(wpercent)))
img = img.resize((basewidth, hsize), Image.ANTIALIAS)

image_data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)

plt.imshow(img)
plt.show()

for i, word in enumerate(image_data['text']):
    if word != "":
        (left, top, right, bottom) = image_data['left'][i], image_data['top'][i], image_data['left'][i] + image_data['width'][i], image_data['top'][i] + image_data['height'][i]
        im_crop = img.crop((left, top, right, bottom))
        plt.imshow(im_crop)
        plt.show()
```
